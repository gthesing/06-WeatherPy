```python
# Dependencies

import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from api_keys import api_key
from citipy import citipy
```

```python
df = pd.read_csv('output_data/cities.csv')
df.head()
```

```python
df.shape
```

```python
# Scatter plot of Temp vs Lat

plt.scatter(df['latitude'], df['temperature'], marker="o")
plt.title('Temperature vs Latitude')
plt.ylabel('Temperature (F)')
plt.xlabel('Latitude')
plt.grid(True)
plt.savefig('Images/TemperatureVsLatitude.png')
plt.show()
```

```python
# Scatter plot of Humidity vs Lat

plt.scatter(df['latitude'], df['humidity'], marker="o")
plt.title('Humidity vs Latitude')
plt.ylabel('Percent Humidity')
plt.xlabel('Latitude')
plt.grid(True)
plt.savefig('Images/HumidityVsLatitude.png')
plt.show()
```
